```python
import autograd.numpy as np
from autograd import grad
from autograd.numpy import log, sqrt, sin, cos, exp, pi, prod
from autograd.numpy.random import normal, uniform

import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
import imageio
import os
from google.colab import drive
drive.mount('/content/drive')

path = "/content/drive/My Drive"

os.chdir(path)
os.listdir(path)
```

```python
np.set_printoptions(precision=3)
np.set_printoptions(suppress=True)
np.random.seed(2021)
```

```python
def mixture(x):
    energy = ((x[0]**2 + x[1]**2)/10 - (cos(2.0*pi*x[0]) + cos(2.0*pi*x[1]))) / 0.5 # 2
    regularizer = ((x[0]**2 + x[1]**2) > 20) * ((x[0]**2 + x[1]**2) - 20)
    return energy + regularizer

def mixture_expand(x, y): return mixture([x, y])
def function_plot(x, y): return np.exp(-mixture([x, y]))

lower, upper = -2.5, 2.5
axis_x = np.linspace(lower, upper, 500)
axis_y = np.linspace(lower, upper, 500)
axis_X, axis_Y = np.meshgrid(axis_x, axis_y)

energy_grid = mixture_expand(axis_X, axis_Y)
prob_grid = function_plot(axis_X, axis_Y)
```

```python
class Sampler:
    def __init__(self, f=None, dim=None, xinit=None, lr=0.1, T=1.0, M = 3, total_epoch = 4e5):
        self.f = f
        self.dim = dim
        self.T = T
        self.M = M
        self.total_epoch = total_epoch
        self.lr0 = lr
        self.lr = lr
      
         
        # initialization for SGLD
        self.x = np.array(xinit)
        

    def stochastic_grad(self, beta):  return grad(self.f)(beta) + 0.25*normal(size=self.dim)

    def stochastic_f(self, beta): return self.f(beta.tolist()) + 0.25*normal(size=1)

    def adjust_learning_rate(self, iter):
      cos_inner = np.pi * (iter % (self.total_epoch // self.M))
      cos_inner /= self.total_epoch // self.M
      cos_out = np.cos(cos_inner) + 1
      lr = 0.5*cos_out*self.lr0
      return lr

    def cycsgld_step(self, iter): 
        self.lr = self.adjust_learning_rate(iter)
        self.x = self.x - self.lr * self.stochastic_grad(self.x) + sqrt(2. * self.lr * self.T) * normal(size=self.dim) 
```

```python
nepoch = 2e5
sampler = Sampler(f=mixture, dim=2, xinit=[0.,0.], lr=1e-2, T=1, M = 10, total_epoch = nepoch)
```

```python
warm_up = 5000
cycsgld_x = np.array([sampler.x])
cycsgld_epoch_path = []
lr_thres = 0.009
```

```python
my_images3 = []
PATH = 'anonymous/Contour-Stochastic-Gradient-Langevin-Dynamics/cyclicalSGLD_result/'
if not os.path.isdir(PATH):
    try:
        os.makedirs(PATH)
    except OSError as exc:  # Python >2.5
        if exc.errno == errno.EEXIST and os.path.isdir(PATH):
            pass
        else:
            raise

for iters in range(int(nepoch)):
    sampler.cycsgld_step(iters)
    if iters > warm_up:
        if iters % 20 == 0:
            if sampler.lr < lr_thres:
              cycsgld_x = np.vstack((cycsgld_x, sampler.x))
              cycsgld_epoch_path.append(iters)
        if iters % 5000 == 0:
              fig3 = plt.figure(figsize=(4, 4))
              plt.contour(axis_X, axis_Y, prob_grid, 10)
              plt.yticks([-4, -2, 0, 2, 4]) 
              plt.scatter(cycsgld_x[:,0], cycsgld_x[:,1], marker='.', s=3, color='k', label="Iteration="+str(iters)) 
              plt.legend(loc="upper left", prop={'size': 10})
              plt.xlim([lower, upper])
              plt.ylim([lower, upper])
              # plt.tight_layout()
              plt.show()
              fig3.canvas.draw()
              image3 = np.frombuffer(fig3.canvas.tostring_rgb(), dtype='uint8').reshape(fig3.canvas.get_width_height()[::-1] + (3,))
              my_images3.append(image3)
              plt.close('all')
          

imageio.mimsave(PATH+'cyclicalSGLD_contour_'+str(sampler.lr0)+'_'+str(sampler.T)+'.gif', my_images3, fps=50)      
# imageio.mimsave(PATH+'CSGLD_PDF'+str(zeta)+"_"+str(psampler.lr)+"_"+str(psampler.T)+'.gif', my_images4, fps=50)

# save the sampels
import pickle
f = open(PATH + 'cyclicalSGLD_samples.txt', 'wb')
pickle.dump([cycsgld_epoch_path ,cycsgld_x], f)
f.close()
```
